```python
# Import neccessary libraries
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline #the plots are displayed inline within the notebook
import seaborn as sns

# split a dataset into training and testing sets for model evaluation
from sklearn.model_selection import train_test_split

# Import model traning libraries
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC

from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score

import warnings
warnings.filterwarnings("ignore") #ignore warnings
```

```python
df = pd.read_csv("creditcard.csv")
```

```python
x = df.drop(columns='Class', axis=1)
y = df['Class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.6, stratify=y)
```

```python
print("x_train shape: ",x_train.shape)
print("x_test shape: ",x_test.shape)
print("y_train shape: ",y_train.shape)
print("y_test shape: ",y_test.shape)
```

# Logistic Regression

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
print("Model Accuracy:", accuracy)
print("Confusion Matrix:\n", conf_matrix)
```

# Support Vector Machine 

```python
model = SVC()
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
print("Model Accuracy:", accuracy)
print("Confusion Matrix:\n", conf_matrix)
```

```python
confusion_matrices = {
    "Logistic Regression": [[170515, 75], [122, 173]],
    "Support Vector Machine": [[170590, 0], [295, 0]],
    "Random Forest": [[170576, 14], [72, 223]],
    "KNN": [[170590, 0], [287, 8]],
    "Decision Tree": [[170510, 80], [83, 212]]
}
```

```python
for model, cm in confusion_matrices.items():
    TN, FP = cm[0]
    FN, TP = cm[1]

    precision = TP / (TP + FP) if (TP + FP) > 0 else 0
    recall = TP / (TP + FN) if (TP + FN) > 0 else 0
    f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

    print(f"Model: {model}")
    print(f"  Precision: {precision:.4f}")
    print(f"  Recall: {recall:.4f}")
    print(f"  F1-Score: {f1:.4f}")
    print()
```
